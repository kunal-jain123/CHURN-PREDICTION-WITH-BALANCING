```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('customer_churn.csv')
```

```python
data.drop(['customerID'],axis=1,inplace=True)
```

```python
data['TotalCharges'].values
```

```python
pd.to_numeric(data['TotalCharges'],errors='coerce').isnull().sum()
```

```python
data[pd.to_numeric(data['TotalCharges'],errors='coerce').isnull()]
```

```python
data = data[data['TotalCharges']!=' ']
```

```python
data.shape
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
categorical = []
for i in data:
    if(data[i].dtypes=='object'):
        categorical.append(i)
```

```python
categorical
```

```python
data[categorical]
```

```python
for i in categorical:
    print(i,'=>',data[i].unique())
```

```python
data.replace('No internet service','No',inplace=True)
data.replace('No phone service','No',inplace=True)
```

```python
for i in categorical:
    print(i,'=>',data[i].unique())
```

```python
yesorno = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
yesorno
```

```python
data[yesorno].head()
```

```python
for i in yesorno:
    data[i].replace({'Yes':1,'No':0},inplace=True)
```

```python
data[yesorno].head()
```

```python
data['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
data = pd.get_dummies(data=data,columns=['InternetService','Contract','PaymentMethod'])
```

```python
data.shape
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
data[cols_to_scale] = scaler.fit_transform(data[cols_to_scale])
```

```python
data.head()
```

```python
x = data.drop(['Churn'],axis=1)
```

```python
y = data['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Dense(20,input_shape=(26,),activation='relu'),
                          keras.layers.Dense(20,activation='relu'),
#                           keras.layers.Dense(10,activation='relu'),
                          keras.layers.Dense(1,activation='sigmoid'),])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
pred = model.predict(x_test)
```

```python
y_pred = []
for i in pred:
    if(i>0.5):
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
